```python
data=pd.read_csv("deliveries.csv")
data.head()
```

```python
runs=data.groupby("batsman")
# runs["batsman_runs"].sum().sort_values(ascending=False).head()
# data[data["batsman_runs"]==4].groupby("batsman")["batsman"].count().sort_values(ascending=False)
data[data["batsman"]=="V Kohli"].groupby("bowling_team")["batsman_runs"].sum().sort_values(ascending=False).head()
```

```python
def max_runs_against_a_team(batsman):
   return data[data["batsman"]==batsman].groupby("bowling_team")["batsman_runs"].sum().sort_values(ascending=False).index[0]
max_runs_against_a_team("MS Dhoni")    
```

```python
# data[data["over"]>15].groupby("batsman")["batsman_runs"].count()>200
death_overs=data[data["over"]>15]
all_batsman=death_overs.groupby("batsman")["batsman_runs"].count()
x=all_batsman>200
listt=all_batsman[x].index.to_list()
final=data[data["batsman"].isin(listt)]
runs=final.groupby("batsman")['batsman_runs'].sum()
balls=final.groupby("batsman")['batsman_runs'].count()
sr=(runs/balls)*100
sr
```
