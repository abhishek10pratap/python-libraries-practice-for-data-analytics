#MATPLOTLIB


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
#2D line plot
#bivariant analysis (takes min 2 columns )
# categorical ->numerical and numeriacal->numerical
# Use case -time series data
price=[40000,50000,47000,51000,36000,49000]
years=[2015,2016,2017,2018,2019,2020]
plt.plot(years,price)
```

```python
batsman=pd.read_csv("sharma-kohli.csv")
plt.plot(batsman["index"],batsman["V Kohli"],color="green")
plt.plot(batsman["index"],batsman["RG Sharma"],color="black")
plt.title("Rohit Sharma vs V Kohli Career Comparison")
plt.xlabel("Seasons")
plt.ylabel("Runs Scored")
```

```python
plt.plot(batsman["index"],batsman["V Kohli"],color="green",linestyle="solid",linewidth=3)
plt.plot(batsman["index"],batsman["RG Sharma"],color="black",linestyle="dashdot",linewidth=4)
plt.title("Rohit Sharma vs V Kohli Career Comparison")
plt.xlabel("Seasons")
plt.ylabel("Runs Scored")
```

```python
plt.plot(batsman["index"],batsman["V Kohli"],color="green",linestyle="solid",linewidth=3,marker="o",markersize=10)
plt.plot(batsman["index"],batsman["RG Sharma"],color="black",linestyle="solid",linewidth=4,marker="o",markersize=12)
plt.title("Rohit Sharma vs V Kohli Career Comparison")
plt.xlabel("Seasons")
plt.ylabel("Runs Scored")
```

```python
plt.plot(batsman["index"],batsman["V Kohli"],color="yellow",linestyle="solid",linewidth=3,marker="o",markersize=10,label="V Kohli")
plt.plot(batsman["index"],batsman["RG Sharma"],color="purple",linestyle="solid",linewidth=4,marker="o",markersize=12,label="R Sharma")
plt.title("Rohit Sharma vs V Kohli Career Comparison")
plt.legend()
plt.xlabel("Seasons")
plt.ylabel("Runs Scored")
plt.grid()
```

```python
#SCATTER PLOT 
# -bivariant 
# numerical vs numerical
# Use case -Finding correlation
x=np.linspace(-10,10,50)
y=10*x+3+np.random.randint(0,300,50)
plt.scatter(x,y)
```

```python
data=pd.read_csv("batter.csv")
df=data.head(50)
plt.scatter(df["avg"],df["strike_rate"],color="red",marker="+")
plt.title("Avg and SR analysis of Top 50 Batsman")
plt.xlabel("Runs scored by batsman")
plt.ylabel("Strike rate of batsman")
```

```python
batters=pd.read_csv("batter.csv")
sample_df=batters.head(100).sample(25,random_state=5)
sample_df
```

```python

plt.figure(figsize=(18,10))
plt.scatter(sample_df["avg"],sample_df["strike_rate"],s=sample_df["runs"])
for i in range(sample_df.shape[0]):
    plt.text(sample_df["avg"].values[i],sample_df["strike_rate"].values[i],sample_df["batter"].values[i])
plt.axhline(130,color="brown") # for more filtering the players based in SR and average
plt.axhline(140,color="blue")
plt.axvline(30,color="red")
```

```python
fig,ax=plt.subplots(figsize=(15,6))
ax.scatter(sample_df["avg"],sample_df["strike_rate"])
ax.set_title("Something")
ax.set_xlabel("Avg")
ax.set_ylabel("strike_rate")
```
